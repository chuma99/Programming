```python
import numpy as np
import matplotlib.pyplot as plt

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 #0 to 15 point radiuses
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

```

```python
import seaborn as sns
sns.set_style("ticks")
tips = 'tips.csv'
tips1 = pd.read_csv(tips) 
g = sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips1);
```

```python
import seaborn as sns
sns.set_style("ticks")
tips = 'tips.csv'
tips1 = pd.read_csv(tips) 
g = sns.relplot(x="total_bill", y="tip", hue="day", data=tips1);
```

```python
import seaborn as sns
sns.set_style("ticks")
tips = 'tips.csv'
tips1 = pd.read_csv(tips) 
g = sns.lineplot(x="total_bill", y="tip", hue="day", data=tips1);
```

```python
import seaborn as sns
sns.set_style("ticks")
tips = 'tips.csv'
tips1 = pd.read_csv(tips) 
g = sns.boxplot(x="total_bill", y="tip", hue="day", data=tips1);
```
